# Convert and filter PAW pipeline results

### Import libraries

```python
import numpy as np
import pandas as pd
import glob

import sys
import os
import re
```

```python
PAW_RESULTS_PATH = '../Initial files/grouped_protein_summary_TMT_8_truncated.csv'
SAMPLE_LABEL_PATH = '../Initial files/TMT-label_sample_key.csv'

# glob.glob(PAW_RESULTS_PATH)
result_df = pd.read_csv(PAW_RESULTS_PATH)
label_df = pd.read_csv(SAMPLE_LABEL_PATH)
label_df
result_df
```

```python
def map_columns(row):
    return '_'.join(['TotInt',row['Tag'],row['Batch']])

label_df['Columns'] = label_df.apply(map_columns,axis=1)

col_dict = dict(zip(label_df['Columns'],label_df['Replicate Name']))
result_df = result_df.rename(columns=col_dict)
result_df
```

```python
COVERAGE_THRESHOLD = 0.0

# Eliminate the small number of contamination protein families
# result_df = result_df[~result_df['Accession'].str.contains("family")]

def find_coverage(num):
    return re.findall(r'[\'\(]*([0-9]+.?[0-9]*)\)?',num)[0]
        
result_df['Coverage'] = result_df['Coverage'].apply(find_coverage)
result_df = result_df.astype({'Coverage': 'float64'})

# result_df['Coverage'] = result_df['Coverage'].apply(convert_str)

filtered_df = result_df[result_df['Coverage'] > COVERAGE_THRESHOLD]
filtered_df
```

```python
# Remove unnecessary columns
eliminate = ['ProtGroup','Counter','Identical','Similar','OtherLoci','Filter','Coverage','SeqLength','MW','CountsTot','UniqueTot','UniqFrac','Total_TMT1','Total_TMT2','Unique_TMT1','Unique_TMT2','Corrected_TMT1','Corrected_TMT2','PSMs_Used_TMT1','PSMs_Used_TMT2']

final_df = filtered_df.drop(columns=eliminate)
float_dict = dict(zip(label_df['Replicate Name'],['float64' for i in range(len(label_df['Replicate Name']))]))
final_df = final_df.astype(float_dict)
```

```python
final_df.to_csv('../Intermediate data/Filtered_PAW_protein_intensities.csv',index=False)
```
